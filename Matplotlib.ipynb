# Formatando o estilo do plot



*   O plot desenha uma axe para cada par de x y
*   Para cada par tem um terceiro argumento opcional que indica a cor e o tipo de plot




```python
plt.plot(t, t, t, t**2, t, t**3, t, t+t*2, t, t+2*t*t)
plt.show()
```

A formatação do plot é formado pela letra que indica a cor e a string do símbolo


```python
plt.plot(t, t, 'bo' , t, t**2, 'g+', t, t**3, 'r--', t, t+t*2, 'r^', t, t+2*t*t, 'g-')
plt.show()
```

# Plotando variáveis categóricas

```python
viaturas = ['AXS', 'Corola', 'Etios']
quantidades = [48, 72, 23]
```

```python
plt.figure(figsize=(10,5))

plt.subplot(131) # O número indica 1 de 3
plt.bar(viaturas, quantidades)
plt.title('Bar')
plt.subplot(132) # O número indica 2 de 3
plt.scatter(viaturas, quantidades)
plt.title('Scatter')
plt.subplot(133) # O número indica 3 de 3
plt.plot(viaturas, quantidades)
plt.title('Subplot')
plt.suptitle('Brasil')
plt.show()
```

O mesmo gráfico usando boas práticas

```python
fig = plt.figure(figsize=(10,5))

ax1 = plt.subplot(1,3,1) # nLin, nCol, nIndex
ax1.bar(viaturas, quantidades)
ax1.set_title("Bar")
ax2 = plt.subplot(132) # O número indica 2 de 3
ax2.scatter(viaturas, quantidades)
ax2.set_title('Scatter')
ax3 = plt.subplot(133) # O número indica 3 de 3
ax3.plot(viaturas, quantidades)
ax3.set_title('Subplot')

fig.suptitle('Título Superior')
fig.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

plt.figure(figsize=(8,6))
plt.subplot(211)
plt.title('Gráfico 1')
plt.plot(t1, f(t1), 'bo', t1, f(t1), 'k--')
plt.subplot(212)
plt.title('Gráfico 2')
plt.plot(t2,f(t2), 'r--')

plt.show()
```

## Anotações

```python
ax = plt.subplot()

t = np.arange(0, 5, 0.01)
s = np.cos(2*np.pi*t)
line = plt.plot(t, s, lw=2)

plt.annotate('máximo', xy=(2,1), xytext=(3, 1.5),
             arrowprops=dict(color='k', shrink=0.05))
plt.annotate('mínimo', xy=(2.5,-1), xytext=(2, -1.8),
             arrowprops=dict(color='r', arrowstyle='<-'))

plt.ylim(-2,2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook

# Load a numpy structured array from yahoo csv data with fields date, open,
# close, volume, adj_close from the mpl-data/example directory.  This array
# stores the date as an np.datetime64 with a day unit ('D') in the 'date'
# column.
data = cbook.get_sample_data('goog.npz', np_load=True)['price_data']

fig, ax = plt.subplots()
ax.plot('date', 'adj_close', data=data)

# Major ticks every 6 months.
fmt_half_year = mdates.MonthLocator(interval=6)
ax.xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax.xaxis.set_minor_locator(fmt_month)

# Text in the x axis will be displayed in 'YYYY-mm' format.
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# Round to nearest years.
datemin = np.datetime64(data['date'][0], 'Y')
datemax = np.datetime64(data['date'][-1], 'Y') + np.timedelta64(1, 'Y')
ax.set_xlim(datemin, datemax)

# Format the coords message box, i.e. the numbers displayed as the cursor moves
# across the axes within the interactive GUI.
ax.format_xdata = mdates.DateFormatter('%Y-%m')
ax.format_ydata = lambda x: f'${x:.2f}'  # Format the price.
ax.grid(True)

# Rotates and right aligns the x labels, and moves the bottom of the
# axes up to make room for them.
fig.autofmt_xdate()

plt.show()
```
